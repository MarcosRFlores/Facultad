Funcion de membresia de un numero difuso triangular

$A(x)=\left\{ \begin{array}{ll}

1+\frac{x-\overline x}{e_1} & \text{ si } \overline x-e_1< x <\overline x\\

Para el primer intervalo:

$1+\frac{x-\overline x}{e_1} \geq α$

Para el otro intervalo

$\overline x\leq x < \overline x+e_R$

$1-\frac{x-\overline x}{e_r}\geq α$

1-\frac{x-2}{5}& \text{ si } 2\leq x < 7\\

\end{array}\right.

3. Dibujo de $tfn$

```python
import numpy as np
import matplotlib.pyplot as plt

def membership_function(x):
    if x <= -2:
        return 0
    elif -2 < x < 2:
        return 1 + (x - 2) / 4
    elif 2 <= x < 7:
        return 1 - (x - 2) / 5
    else:
        return 0

x_vals = np.linspace(-5, 10, 400)  
y_vals = [membership_function(x) for x in x_vals]  

plt.plot(x_vals, y_vals)
plt.title("Función de Membresía")
plt.xlabel("x")
plt.ylabel("Valor de membresía")
plt.grid(True)
plt.show()
```

1. Escribe la función de membresía del número.
2. Demuestra que sus cortes alfa están dados por $A_\alpha=[\sqrt r,3-2\sqrt r]$.
3. Grafica el número.

1. Si tenemos que   $L(x)=R(x)=x^2$. Sean $a_1^-=a_1^+=1$ y $\underline{a}=1$, $\overline{a}=2$

podemos calcular $a_0^-$ y   $a_0^+$ 

sabemos que $\underline{a}=a_1^--a_0^-$ $-> 1=1-a_0^-=a_0^-=0$

sabemos que $\overline{a}=a_0^+-a_1^+$ $-> 2=a_0^+-1=a_0^+-=3$

Con esto podemos escribir la funciona de membresia 


$A(x)=\left\{ \begin{array}{ll}
0             & \text{ si } x<0 \\
L(\frac{x-0}{1}) & \text{ si } 0 \leq x <1 \\
1 & \text{ si } 1 \leq x <1 \\
R(\frac{3-x}{2}) & \text{ si } 1 \leq x < 3\\
0             & \text{ si } x\geq 3 ,
\end{array}\right.
$

2.

La formula para obtener los cortes alfas estan dados por $u_r=[a_0^-+L^{-1}(r)\underline{a},a_0^+-R^{-1}(r)\overline{a}]$

Remplanzando los valores obtenidos anteriormente $u_r=[0^-+\sqrt{r}1,3-\sqrt{r}2]$ $--->u_r=[\sqrt{r},3-\sqrt{r}2]$

```python
import numpy as np
import matplotlib.pyplot as plt

def L(x):
    return x**2

def R(x):
    return x**2

def A(x):
    if x < 0:
        return 0
    elif 0 <= x < 1:
        return L((x - 0) / 1)
    elif 1 <= x < 2:
        return 1
    elif 2 <= x < 3:
        return R((3 - x) / 2)
    else:
        return 0

x_vals = np.linspace(-1, 4, 1000)
y_vals = np.array([A(x) for x in x_vals])

plt.plot(x_vals, y_vals, label='$A(x)$')
plt.xlabel('x')
plt.ylabel('A(x)')
plt.title('Gráfico de la función $A(x)$')
plt.axhline(y=0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(x=0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(x=1, color='black', linewidth=0.8, linestyle='--')
plt.axvline(x=2, color='black', linewidth=0.8, linestyle='--')
plt.axvline(x=3, color='black', linewidth=0.8, linestyle='--')
plt.legend()
plt.grid()
plt.show()
```

**Ejercicio 7.** Sea $u=\text{trfn}(2,5,6,7)$ un número difuso trapezoidal y $v=\text{tfn}(1,4,5)$ un número difuso triangular. Dibuja $u$ y $v$ y calcula la función de membresía de:

La función de membresía de la suma $u + v$ se calcula sumando las funciones de membresía $\mu_u(x)$ y $\mu_v(x)$:

- Si $x \leq 1$: $\mu_{u+v}(x) = 0$
- Si $1 \leq x \leq 2$: $\mu_{u+v}(x) = \frac{x - 1}{4 - 1} + \frac{x - 2}{5 - 2}$
- Si $2 \leq x \leq 4$: $\mu_{u+v}(x) = \frac{x - 2}{5 - 2} + \frac{x - 2}{5 - 2}$

- Si $5 \leq x \leq 6$: $\mu_{u+v}(x) = \frac{7 - x}{7 - 6}$
- Si $6 \leq x \leq 7$: $\mu_{u+v}(x) = \frac{7 - x}{7 - 6} + \frac{7 - x}{7 - 6}$
- Si $x \geq 7$: $\mu_{u+v}(x) = 0$

-Función de membresía de u - v

La función de membresía de la diferencia \( u - v \) se calcula restando las funciones de membresía $\mu_u(x)$ y  $\mu_v(x)$:

- Si $x \leq 1$: $\mu_{u-v}(x) = 0$
- Si $1 \leq x \leq 2$: $\mu_{u-v}(x) = \frac{x - 1}{4 - 1} - \frac{x - 1}{5 - 1}$
- Si $2 \leq x \leq 4$: $\mu_{u-v}(x) = \frac{x - 1}{4 - 1} - \frac{x - 2}{5 - 2}$
- Si $4 \leq x \leq 5$: $\mu_{u-v}(x) = \frac{x - 1}{4 - 1} - 0$
- Si $5 \leq x \leq 6$: $\mu_{u-v}(x) = 0$
- Si $6 \leq x \leq 7$: $\mu_{u-v}(x) = 0$
- Si $x \geq 7$: $\mu_{u-v}(x) = 0$


-Función de membresía de -2u + v

La función de membresía de la operación \( -2u + v \) se calcula restando el doble de la función de membresía $\mu_u(x)$ y la función de membresía $\mu_v(x)$:

- Si $x \leq 1$:$\mu_{-2u+v}(x) = 0$
- Si $1 \leq x \leq 2$: $\mu_{-2u+v}(x) = -2 \cdot \frac{x - 1}{4 - 1} + \frac{x - 2}{5 - 2}$
- Si $2 \leq x \leq 4$: $\mu_{-2u+v}(x) = -2 \cdot \frac{x - 2}{5 - 2} + \frac{x - 2}{5 - 2}$
- Si $4 \leq x \leq 5$: $\mu_{-2u+v}(x) = -2 \cdot \frac{x - 2}{5 - 2} + \frac{5 - x}{5 - 4}$
- Si $5 \leq x \leq 6$: $\mu_{-2u+v}(x) = -2 \cdot \frac{7 - x}{7 - 6}$
- Si $6 \leq x \leq 7$: $\mu_{-2u+v}(x) = -2 \cdot \frac{7 - x}{7 - 6} + \frac{7 - x}{7 - 6}$
- Si $x \geq 7$: $\mu_{-2u+v}(x) = 0$


-Función de membresía de u - u

La función de membresía de la operación \( u - u \) se calcula restando la función de membresía $\mu_u(x)$ a sí misma:

- Si $x \leq 1$: $\mu_{u-u}(x) = 0$
- Si $1 \leq x \leq 2$: $\mu_{u-u}(x) = 0$
- Si $2 \leq x \leq 4$: $\mu_{u-u}(x) = 0$
- Si $4 \leq x \leq 5$: $\mu_{u-u}(x) = 0$
- Si $5 \leq x \leq 6$: $\mu_{u-u}(x) = 0$
- Si $6 \leq x \leq 7$: $\mu_{u-u}(x) = 0$
- Si $x \geq 7$: $\mu_{u-u}(x) = 0$


-Función de membresía de 2v - v

La función de membresía de la operación \( 2v - v \) se calcula restando la función de membresía $\mu_v(x)$ a sí misma dos veces:

- Si $x \leq 1$: $\mu_{2v-v}(x) = 0$
- Si $1 \leq x \leq 2$: $\mu_{2v-v}(x) = 0$
- Si $2 \leq x \leq 4$: $\mu_{2v-v}(x) = 0$
- Si $4 \leq x \leq 5$: $\mu_{2v-v}(x) = 0$
- Si $5 \leq x \leq 6$: $\mu_{2v-v}(x) = 0$
- Si $6 \leq x \leq 7$: $\mu_{2v-v}(x) = 0$
- Si $x \geq 7$: $\mu_{2v-v}(x) = 0$


```python
import numpy as np
import matplotlib.pyplot as plt

# Función trapezoidal corregida para evitar división por cero
def trapezoidal(x, a1, a2, a3, a4):
    # Evitar división por cero con un pequeño valor epsilon
    epsilon = 1e-9
    return np.maximum(0, np.minimum((x - a1) / (max(a2 - a1, epsilon)), np.minimum(1, (a4 - x) / (max(a4 - a3, epsilon)))))

# Función triangular
def triangular(x, b1, b2, b3):
    return np.maximum(0, np.minimum((x - b1) / (b2 - b1), (b3 - x) / (b3 - b2)))

# Definir los números difusos
x = np.linspace(-20, 20, 500)
u = trapezoidal(x, 2, 5, 6, 7)
v = triangular(x, 1, 4, 5)
u_plus_v = trapezoidal(x, 3, 9, 11, 12)
u_minus_v = trapezoidal(x, -3, 1, 2, 6)
neg_2u_plus_v = trapezoidal(x, -13, -8, -5, 1)
u_minus_u = trapezoidal(x, 0, 0, 0, 0)  # Esto será una línea plana en 0
two_v_minus_v = triangular(x, 1, 4, 5)

# Graficar los resultados
plt.plot(x, u, label='u = trfn(2, 5, 6, 7)')
plt.plot(x, v, label='v = tfn(1, 4, 5)')
plt.plot(x, u_plus_v, label='u + v')
plt.plot(x, u_minus_v, label='u - v')
plt.plot(x, neg_2u_plus_v, label='-2u + v')
plt.plot(x, u_minus_u, label='u - u')
plt.plot(x, two_v_minus_v, label='2v - v')

plt.title('Resultados de operaciones difusas')
plt.xlabel('x')
plt.ylabel('Membresía')
plt.legend()
plt.grid(True)
plt.show()
```

1. Para cualquier número difuso $u$ y cualquier $a,b\in \mathbb{R}$, con $a\cdot b\geq 0$, se cumple $(a+b)\cdot u=(a\cdot u)+(b\cdot u)$.
2. Para cualquier $\lambda\in\mathbb{R}$ y cualesquiera números difusos $u,v$, se cumple $\lambda\cdot (u+v)=\lambda\cdot u+\lambda\cdot v$.
3. Para cualquier $\lambda,\mu\in\mathbb{R}$ y cualquier número difuso $u$, se cumple $(\lambda\cdot \mu)\cdot u=\lambda\cdot (\mu\cdot u)$.

1. Para cualquier número difuso u y cualquier $a, b ∈ ℝ$, con $a · b ≥ 0$, se cumple $(a + b) · u = (a · u) + (b · u)$
Demostración:
Sea u un número difuso representado como una función $u(x): ℝ → [0, 1]$ y $a, b ∈ ℝ$. Si $a · b ≥ 0$, entonces se cumple la propiedad distributiva de los números reales:

    $(a + b) · u(x) = a · u(x) + b · u(x)$

    Esto demuestra que:

    $(a + b) · u = (a · u) + (b · u)$

2. Demostración:
Sea $u(x) y v(x)$ números difusos. El lado izquierdo de la propiedad implica multiplicar $λ$ por la suma de los números difusos $u y v$:

    $λ · (u(x) + v(x)) = λ · u(x) + λ · v(x)$

    Por lo tanto, la propiedad distributiva del producto escalar se cumple:

    $λ · (u + v) = λ · u + λ · v$

3. Demostración:
Sea $u(x)$ un número difuso. El lado izquierdo de la ecuación representa el producto escalar entre $(λ · μ)$ y $u(x)$, mientras que el lado derecho representa la multiplicación en dos pasos:

    $(λ · μ) · u(x) = λ · (μ · u(x))$

    Ambos lados de la igualdad son equivalentes debido a la propiedad asociativa del producto en los números reales:

    $(λ · μ) · u = λ · (μ · u)$

0 & \text{si } x \leq 2 \\

1 & \text{si } 5 < x \leq 6 \\
\frac{8 - x}{8 - 6} & \text{si } 6 < x \leq 8 \\
0 & \text{si } x \geq 8
\end{cases}
$$

Para el número triangular $v = \text{tfn}(1, 4, 5)$, su función de membresía es:

$$
\mu_v(x) =
\begin{cases} 
0 & \text{si } x \leq 1 \\
\frac{x - 1}{4 - 1} & \text{si } 1 < x \leq 4 \\
\frac{5 - x}{5 - 4} & \text{si } 4 < x \leq 5 \\
0 & \text{si } x \geq 5
\end{cases}
$$

Para calcular el producto $u \cdot v$, multiplicamos las funciones de membresía $\mu_u(x)$ y $\mu_v(x)$ en cada punto.

Los valores extremos son:

$a_1 = 2$, $a_2 = 5$, $a_3 = 6$, $a_4 = 8$ para $u$.  
$b_1 = 1$, $b_2 = 4$, $b_3 = 5$ para $v$.

El intervalo del corte alfa será:

$$
[\alpha_L, \alpha_U] = [\min(a_1 \cdot b_1, a_1 \cdot b_3, a_4 \cdot b_1, a_4 \cdot b_3), \max(a_2 \cdot b_2, a_3 \cdot b_2)]
$$

$$
[\alpha_L, \alpha_U] = [\min(2 \cdot 1, 2 \cdot 5, 8 \cdot 1, 8 \cdot 5), \max(5 \cdot 4, 6 \cdot 4)]
$$

$$
[\alpha_L, \alpha_U] = [\min(2, 10, 8, 40), \max(20, 24)] = [2, 24]
$$

Por lo tanto, el intervalo del corte alfa de $u \cdot v$ es $[2, 24]$.

```python
import numpy as np
import matplotlib.pyplot as plt

def trapezoidal(x, a1, a2, a3, a4):
    return np.maximum(0, np.minimum((x - a1) / (a2 - a1), np.minimum(1, (a4 - x) / (a4 - a3))))

def triangular(x, b1, b2, b3):
    return np.maximum(0, np.minimum((x - b1) / (b2 - b1), (b3 - x) / (b3 - b2)))

x = np.linspace(0, 50, 500)
u = trapezoidal(x, 2, 5, 6, 8)
v = triangular(x, 1, 4, 5)

u_v = u * v

plt.plot(x, u, label='u = trfn(2, 5, 6, 8)')
plt.plot(x, v, label='v = tfn(1, 4, 5)')
plt.plot(x, u_v, label='u * v', linestyle='--')

plt.title('Gráfica de $u$, $v$ y $u * v$')
plt.xlabel('x')
plt.ylabel('Membresía')
plt.legend()
plt.grid(True)
plt.show()
```
