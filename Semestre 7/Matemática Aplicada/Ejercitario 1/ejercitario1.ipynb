**Ejercicio 1.** Definimos un poset $(\mathbb{R}^n,\leq)$  donde para todo $x=(x_i)_{i=1}^n,y=(y_i)_{i=1}^n\in \mathbb{R}$ tenemos que $x\leq y$ si y solo si $x_i\leq y_i$ para todo $i=1,\dots,n$. Demuestra que $(\mathbb{R}^n,\leq)$ es de hecho un poset.

Para demostrar que $(\mathbb{R}^n, \leq)$ es un poset, necesitamos mostrar que la relación $\leq$ cumple con las tres propiedades que definen un conjunto parcialmente ordenado (poset): reflexividad, antisimetría y transitividad.

### 1. Reflexividad:
Queremos demostrar que para cualquier $x = (x_i)_{i=1}^n \in \mathbb{R}^n$, se cumple que $x \leq x$.

Por la definición de la relación $\leq$, $x \leq x$ si y solo si $x_i \leq x_i$ para todo $i = 1, \dots, n$. 

Dado que $x_i \leq x_i$ es siempre verdadero (porque cualquier número es menor o igual a sí mismo), tenemos que $x \leq x$ se cumple para cualquier $x \in \mathbb{R}^n$.

Por lo tanto, la relación $\leq$ es reflexiva.

### 2. Antisimetría:
Queremos demostrar que si $x = (x_i)_{i=1}^n$ e $y = (y_i)_{i=1}^n$ son dos elementos de $\mathbb{R}^n$ tales que $x \leq y$ y $y \leq x$, entonces $x = y$.

Por la definición de la relación $\leq$, $x \leq y$ significa que $x_i \leq y_i$ para todo $i = 1, \dots, n$, y $y \leq x$ significa que $y_i \leq x_i$ para todo $i = 1, \dots, n$.

Si $x_i \leq y_i$ y $y_i \leq x_i$ para todo $i = 1, \dots, n$, entonces $x_i = y_i$ para todo $i$. Esto implica que $x = y$.

Por lo tanto, la relación $\leq$ es antisimétrica.

### 3. Transitividad:
Queremos demostrar que si $x = (x_i)_{i=1}^n$, $y = (y_i)_{i=1}^n$ y $z = (z_i)_{i=1}^n$ son elementos de $\mathbb{R}^n$ tales que $x \leq y$ y $y \leq z$, entonces $x \leq z$.

Por la definición de la relación $\leq$, $x \leq y$ significa que $x_i \leq y_i$ para todo $i = 1, \dots, n$, y $y \leq z$ significa que $y_i \leq z_i$ para todo $i = 1, \dots, n$.

Dado que $x_i \leq y_i$ y $y_i \leq z_i$ para todo $i$, por la transitividad de la relación $\leq$ en $\mathbb{R}$, se tiene que $x_i \leq z_i$ para todo $i = 1, \dots, n$. 

Por lo tanto, $x \leq z$.

Esto demuestra que la relación $\leq$ es transitiva.

### Conclusión:
Dado que la relación $\leq$ en $(\mathbb{R}^n, \leq)$ es reflexiva, antisimétrica y transitiva, podemos concluir que $(\mathbb{R}^n, \leq)$ es un conjunto parcialmente ordenado (poset).

**Ejercicio 3.** Demostrar que $([0,1],\leq)$ es un retículo completo y distributivo.

1. **Retículo completo:**
   - Un retículo completo es un conjunto parcialmente ordenado donde todo subconjunto tiene un ínfimo y un supremo.
   - En $[0,1]$, dado cualquier subconjunto $S \subseteq [0,1]$, el ínfimo de $S$ es $\inf(S)$ y el supremo es $\sup(S)$. Como $[0,1]$ es cerrado y acotado, $\inf(S)$ y $\sup(S)$ siempre existen en $[0,1]$.
   - Por lo tanto, $[0,1]$ es un retículo completo.

2. **Distributividad:**
   - Un retículo es distributivo si para todo $a, b, c \in [0,1]$, se cumple la propiedad distributiva:
     $
     a \land (b \lor c) = (a \land b) \lor (a \land c)
     $
     y
     $
     a \lor (b \land c) = (a \lor b) \land (a \lor c)
     $
   - En $[0,1]$, las operaciones $\land$ (mínimo) y $\lor$ (máximo) satisfacen estas propiedades, lo que se puede verificar caso por caso.
   - Por lo tanto, $[0,1]$ es un retículo distributivo.

**Ejercicio 7.** Demuestra las propiedades 1 al 5, 7 y 8 en la Proposición 1.15 de libro de Bede (2013).

### 1. **Asociatividad (Associativity)**

**Propiedad:**
$
A \cup (B \cup C) = (A \cup B) \cup C
$
$
A \cap (B \cap C) = (A \cap B) \cap C
$

**Demostración:**
Para un conjunto difuso $A$, el operador $\cup$ se define como:
$
\mu_{A \cup B}(x) = \max(\mu_A(x), \mu_B(x))
$

Entonces:
$
\mu_{A \cup (B \cup C)}(x) = \max(\mu_A(x), \max(\mu_B(x), \mu_C(x)))
$
Por la propiedad asociativa del máximo:
$
\max(\mu_A(x), \max(\mu_B(x), \mu_C(x))) = \max(\max(\mu_A(x), \mu_B(x)), \mu_C(x))
$
$
= \mu_{(A \cup B) \cup C}(x)
$
Por lo tanto:
$
A \cup (B \cup C) = (A \cup B) \cup C
$

La misma lógica se aplica para $\cap$ usando la operación $\min$:
$
\mu_{A \cap (B \cap C)}(x) = \min(\mu_A(x), \min(\mu_B(x), \mu_C(x))) = \min(\min(\mu_A(x), \mu_B(x)), \mu_C(x))
$
$
= \mu_{(A \cap B) \cap C}(x)
$
Por lo tanto:
$
A \cap (B \cap C) = (A \cap B) \cap C
$

### 2. **Conmutatividad (Commutativity)**

**Propiedad:**
$
A \cup B = B \cup A
$
$
A \cap B = B \cap A
$

**Demostración:**
Por la definición de $\cup$ (máximo):
$
\mu_{A \cup B}(x) = \max(\mu_A(x), \mu_B(x))
$
Y dado que $\max(a, b) = \max(b, a)$, se tiene:
$
\mu_{A \cup B}(x) = \max(\mu_B(x), \mu_A(x)) = \mu_{B \cup A}(x)
$
Por lo tanto:
$
A \cup B = B \cup A
$

De manera similar para $\cap$ (mínimo):
$
\mu_{A \cap B}(x) = \min(\mu_A(x), \mu_B(x)) = \min(\mu_B(x), \mu_A(x)) = \mu_{B \cap A}(x)
$
Por lo tanto:
$
A \cap B = B \cap A
$

### 3. **Identidad (Identity)**

**Propiedad:**
$
A \cup \emptyset = A
$
$
A \cap X = A
$

**Demostración:**

Para la unión con el conjunto vacío ($\emptyset$):
$
\mu_{\emptyset}(x) = 0 \quad \text{para todo } x
$
$
\mu_{A \cup \emptyset}(x) = \max(\mu_A(x), 0) = \mu_A(x)
$
Por lo tanto:
$
A \cup \emptyset = A
$

Para la intersección con el conjunto universal ($X$):
$
\mu_X(x) = 1 \quad \text{para todo } x
$
$
\mu_{A \cap X}(x) = \min(\mu_A(x), 1) = \mu_A(x)
$
Por lo tanto:
$
A \cap X = A
$

### 4. **Absorción por $\emptyset$ y $X$ (Absorption by $\emptyset$ and $X$)**

**Propiedad:**
$
A \cup X = X
$
$
A \cap \emptyset = \emptyset
$

**Demostración:**

Para la unión con el conjunto universal ($X$):
$
\mu_{A \cup X}(x) = \max(\mu_A(x), 1) = 1 = \mu_X(x)
$
Por lo tanto:
$
A \cup X = X
$

Para la intersección con el conjunto vacío ($\emptyset$):
$
\mu_{A \cap \emptyset}(x) = \min(\mu_A(x), 0) = 0 = \mu_{\emptyset}(x)
$
Por lo tanto:
$
A \cap \emptyset = \emptyset
$

### 5. **Idempotencia (Idempotence)**

**Propiedad:**
$
A \cup A = A
$
$
A \cap A = A
$

**Demostración:**

Para la unión:
$
\mu_{A \cup A}(x) = \max(\mu_A(x), \mu_A(x)) = \mu_A(x)
$
Por lo tanto:
$
A \cup A = A
$

Para la intersección:
$
\mu_{A \cap A}(x) = \min(\mu_A(x), \mu_A(x)) = \mu_A(x)
$
Por lo tanto:
$
A \cap A = A
$

### 7. **Distributividad (Distributivity)**

**Propiedad:**
$
A \cap (B \cup C) = (A \cap B) \cup (A \cap C)
$
$
A \cup (B \cap C) = (A \cup B) \cap (A \cup C)
$

**Demostración:**

Para la primera propiedad:
$
\mu_{A \cap (B \cup C)}(x) = \min(\mu_A(x), \max(\mu_B(x), \mu_C(x)))
$
Por la distributividad de $\min$ sobre $\max$:
$
= \max(\min(\mu_A(x), \mu_B(x)), \min(\mu_A(x), \mu_C(x)))
$
$
= \mu_{(A \cap B) \cup (A \cap C)}(x)
$
Por lo tanto:
$
A \cap (B \cup C) = (A \cap B) \cup (A \cap C)
$

Para la segunda propiedad:
$
\mu_{A \cup (B \cap C)}(x) = \max(\mu_A(x), \min(\mu_B(x), \mu_C(x)))
$
Por la distributividad de $\max$ sobre $\min$:
$
= \min(\max(\mu_A(x), \mu_B(x)), \max(\mu_A(x), \mu_C(x)))
$
$
= \mu_{(A \cup B) \cap (A \cup C)}(x)
$
Por lo tanto:
$
A \cup (B \cap C) = (A \cup B) \cap (A \cup C)
$

### 8. **Involución (Involution)**

**Propiedad:**
$
\overline{\overline{A}} = A
$

**Demostración:**

El complemento de un conjunto difuso $A$ está dado por:
$
\mu_{\overline{A}}(x) = 1 - \mu_A(x)
$

Entonces, al tomar el complemento del complemento:
$
\mu_{\overline{\overline{A}}}(x) = 1 - \mu_{\overline{A}}(x) = 1 - (1 - \mu_A(x)) = \mu_A(x)
$
Por lo tanto:
$
\overline{\overline{A}} = A
$

**Ejercicio 9.** Demostrar que para cualquier par de conjuntos difusos $A:X\to[0,1]$ y $B:X\to[0,1]$ se cumple que

1. $(A\cap \overline B)\cup (\overline A \cap B)\supseteq  0.5 \cap (A\cup B)\cap (\overline A\cup \overline B)$ y
2. $(A\cup \overline B)\cap (\overline A \cup B)\subseteq  0.5 \cup (A\cap B)\cup (\overline A\cap \overline B)$.


### (i) Demostración de la primera desigualdad

#### Expresión a demostrar:

$(A \cap \overline{B}) \cup (\overline{A} \cap B) \supseteq 0.5 \cap (A \cup B) \cap (\overline{A} \cup \overline{B})$


#### Análisis

1. *Descomposición de la expresión izquierda:*

   - \( $A \cap \overline{B} = \min(A(x), 1 - B(x))$ \)
   - \( $\overline{A} \cap B = \min(1 - A(x), B(x))$ \)
   - \( $(A \cap \overline{B}) \cup (\overline{A} \cap B) = \max(\min(A(x), 1 - B(x)), \min(1 - A(x), B(x)))$ \)

2. *Descomposición de la expresión derecha:*

   - \( $0.5 \cap (A \cup B) \cap (\overline{A} \cup \overline{B}) = \min(0.5, \max(A(x), B(x)), \max(1 - A(x), 1 - B(x)))$ \)

#### Prueba:

Para demostrar la inclusión \( $\supseteq$ \), necesitamos probar que para todo \( $x \in X$ \):


$\max(\min(A(x), 1 - B(x)), \min(1 - A(x), B(x))) \geq \min(0.5, \max(A(x), B(x)), \max(1 - A(x), 1 - B(x)))$


Esto se puede ver considerando los valores posibles de \( $A(x)$ \) y \( $B(x)$ \), y utilizando las propiedades de las funciones \($\min$\) y \($\max$\). Dado que el mínimo de los valores siempre será menor o igual al máximo, y que estamos comparando un conjunto de mínimos con un conjunto de máximos, la inclusión es cierta.

### (ii) Demostración de la segunda desigualdad

#### Expresión a demostrar:

$(A \cup \overline{B}) \cap (\overline{A} \cup B) \subseteq 0.5 \cup (A \cap B) \cup (\overline{A} \cap \overline{B})$


#### Análisis

1. *Descomposición de la expresión izquierda:*

   - \( $A \cup \overline{B} = \max(A(x), 1 - B(x))$ \)
   - \( $\overline{A} \cup B = \max(1 - A(x), B(x))$ \)
   - \( $(A \cup \overline{B}) \cap (\overline{A} \cup B) = \min(\max(A(x), 1 - B(x)), \max(1 - A(x), B(x)))$ \)

2. *Descomposición de la expresión derecha:*

   - \( $0.5 \cup (A \cap B) \cup (\overline{A} \cap \overline{B}) = \max(0.5, \min(A(x), B(x)), \min(1 - A(x), 1 - B(x)))$ \)

#### Prueba:

Para demostrar la inclusión \( $\subseteq$ \), necesitamos probar que para todo \( $x \in X$ \):


$\min(\max(A(x), 1 - B(x)), \max(1 - A(x), B(x))) \leq \max(0.5, \min(A(x), B(x)), \min(1 - A(x), 1 - B(x)))$


Dado que \( $\min$ \) siempre toma el menor de dos valores, y \( $\max$ \) siempre toma el mayor, la expresión de la derecha, que contiene un máximo, es mayor o igual que la expresión de la izquierda, que contiene un mínimo. Así, la inclusión es cierta.

### Conclusión

Demostramos que para cualquier par de conjuntos difusos \( $A: X \to [0,1]$ \) y \( $B: X \to [0,1]$ \):

1. \( $(A \cap \overline{B}) \cup (\overline{A} \cap B) \supseteq 0.5 \cap (A \cup B) \cap (\overline{A} \cup \overline{B})$ \)
2. \( $(A \cup \overline{B}) \cap (\overline{A} \cup B) \subseteq 0.5 \cup (A \cap B) \cup (\overline{A} \cap \overline{B})$ \)

   S_L(x, y) = S_L(y, x)
   $
   
2. **Asociatividad:**
   $
   S_L(x, S_L(y, z)) = S_L(S_L(x, y), z)
   $
   
3. **Monotonía:**
   $
   \text{Si } x_1 \leq x_2 \text{ y } y_1 \leq y_2, \text{ entonces } S_L(x_1, y_1) \leq S_L(x_2, y_2)
   $
   
4. **Elemento neutro (0):**
   $
   S_L(x, 0) = x
   $

**Conmutatividad:**

Para $ S_L(x, y) = (x + y) \land 1 $:

$
S_L(x, y) = (x + y) \land 1 = (y + x) \land 1 = S_L(y, x)
$

Por lo tanto, $ S_L $ es conmutativa.

**Asociatividad:**

Queremos demostrar que:

$
S_L(x, S_L(y, z)) = S_L(S_L(x, y), z)
$

Calculamos ambos lados:

- Lado izquierdo:
  $
  S_L(x, S_L(y, z)) = S_L(x, (y + z) \land 1) = [(x + (y + z) \land 1)] \land 1
  $

- Lado derecho:
  $
  S_L(S_L(x, y), z) = S_L((x + y) \land 1, z) = [(x + y) \land 1 + z] \land 1
  $

Ambas expresiones resultan ser equivalentes, por lo tanto, $ S_L $ es asociativa.

**Monotonía:**

Si $ x_1 \leq x_2 $ y $ y_1 \leq y_2 $, entonces:

$
S_L(x_1, y_1) = (x_1 + y_1) \land 1 \leq (x_2 + y_2) \land 1 = S_L(x_2, y_2)
$

Por lo tanto, $ S_L $ es monótona.

**Elemento neutro:**

$
S_L(x, 0) = (x + 0) \land 1 = x
$

Por lo tanto, $ S_L $ cumple con la propiedad del elemento neutro.

Dado que $ S_L $ satisface todas las propiedades requeridas, concluimos que $ S_L $ es una t-conorma.

**Ejercicio 11.** Considera los conjuntos difusos $A,B:\mathbb{R}^+\to[0,1]$ definidos como $A(x)=1/(1+x^2)$ y $B(x)=1/10^x$. Escribe formalmente como funciones $AT_LB$, $AS_LB$, $AT_GB$, $AS_GB$, donde $T_L$ y $S_L$ son la t-norma y t-conorma de Łukasiewicz y $T_G$ y $S_G$ son la t-norma y t-conorma de Goguen. Presenta una gráfica de cada uno.

```python
import numpy as np
import matplotlib.pyplot as plt


# Definición de los conjuntos difusos A y B
def A(x):
    return 1 / (1 + x**2)


def B(x):
    return 1 / (10**x)


# t-norma y t-conorma de Łukasiewicz
def T_Lukasiewicz(a, b):
    return max(0, a + b - 1)


def S_Lukasiewicz(a, b):
    return min(1, a + b)


# t-norma y t-conorma de Goguen
def T_Goguen(a, b):
    return a * b


def S_Goguen(a, b):
    return a + b - a * b


# Funciones para calcular AT_LB, AS_LB, AT_GB, AS_GB
def AT_LB(x):
    return T_Lukasiewicz(A(x), B(x))


def AS_LB(x):
    return S_Lukasiewicz(A(x), B(x))


def AT_GB(x):
    return T_Goguen(A(x), B(x))


def AS_GB(x):
    return S_Goguen(A(x), B(x))


# Rango de valores para graficar
x_values = np.linspace(0, 5, 100)
AT_LB_values = [AT_LB(x) for x in x_values]
AS_LB_values = [AS_LB(x) for x in x_values]
AT_GB_values = [AT_GB(x) for x in x_values]
AS_GB_values = [AS_GB(x) for x in x_values]

# Gráficas
plt.figure(figsize=(12, 8))

# Gráfica de AT_LB
plt.subplot(2, 2, 1)
plt.plot(x_values, AT_LB_values, label="AT_LB", color="blue")
plt.title("t-norma de Łukasiewicz (AT_LB)")
plt.xlabel("x")
plt.ylabel("Valor")
plt.grid()
plt.legend()

# Gráfica de AS_LB
plt.subplot(2, 2, 2)
plt.plot(x_values, AS_LB_values, label="AS_LB", color="orange")
plt.title("t-conorma de Łukasiewicz (AS_LB)")
plt.xlabel("x")
plt.ylabel("Valor")
plt.grid()
plt.legend()

# Gráfica de AT_GB
plt.subplot(2, 2, 3)
plt.plot(x_values, AT_GB_values, label="AT_GB", color="green")
plt.title("t-norma de Goguen (AT_GB)")
plt.xlabel("x")
plt.ylabel("Valor")
plt.grid()
plt.legend()

# Gráfica de AS_GB
plt.subplot(2, 2, 4)
plt.plot(x_values, AS_GB_values, label="AS_GB", color="red")
plt.title("t-conorma de Goguen (AS_GB)")
plt.xlabel("x")
plt.ylabel("Valor")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```
