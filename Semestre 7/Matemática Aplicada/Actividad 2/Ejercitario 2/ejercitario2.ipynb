**(3) Encuentra una expresión intervalar para los cortes alfa de $ A^2 $.**


Para resolver este ejercicio y encontrar una expresión intervalar para los cortes alfa de $ A^2 $, donde $ A = \text{efn}(2, 0.5, 1, 0.4) $, seguimos los siguientes pasos:

#### Paso 1: Definir el número difuso exponencial

El número difuso exponencial $ A = \text{efn}(2, 0.5, 1, 0.4) $ tiene los siguientes parámetros:

- $ x = 2 $ es el valor medio o central.
- $ \beta_l = 0.5 $ es el parámetro del lado izquierdo (pendiente de decrecimiento exponencial en el lado izquierdo).
- $ \beta_r = 1 $ es el parámetro del lado derecho (pendiente de decrecimiento exponencial en el lado derecho).
- $ \alpha_l = 0.4 $ es el parámetro adicional que modifica el comportamiento en el lado izquierdo.

#### Paso 2: Expresión de los cortes alfa para $ A $

Para un número difuso exponencial $ A = \text{efn}(x, \beta_l, \beta_r) $, los cortes alfa están dados por las siguientes expresiones:

- Extremo izquierdo del corte alfa:
  $
  A_l(\alpha) = x - \frac{1}{\beta_l} \ln(1 - \alpha)
  $
  
- Extremo derecho del corte alfa:
  $
  A_r(\alpha) = x + \frac{1}{\beta_r} \ln(1 - \alpha)
  $

#### Paso 3: Expresión para $ A^2 $

Cuando elevamos $ A $ al cuadrado, necesitamos realizar la operación con los cortes alfa de $ A $. Los cortes alfa de $ A^2 $ están dados por:

$
A^2(\alpha) = [A_l(\alpha)^2, A_r(\alpha)^2]
$

#### Paso 4: Sustitución de las expresiones para $ A_l(\alpha) $ y $ A_r(\alpha) $

Sustituyendo las expresiones de los extremos izquierdo y derecho de $ A $ en la fórmula para $ A^2 $:

- Extremo izquierdo del corte alfa de $ A^2 $:
  $
  A_l^2(\alpha) = \left( 2 - \frac{1}{0.5} \ln(1 - \alpha) \right)^2 = \left( 2 - 2 \ln(1 - \alpha) \right)^2
  $
  
- Extremo derecho del corte alfa de $ A^2 $:
  $
  A_r^2(\alpha) = \left( 2 + \frac{1}{1} \ln(1 - \alpha) \right)^2 = \left( 2 + \ln(1 - \alpha) \right)^2
  $

#### Paso 5: Expresión final para los cortes alfa de $ A^2 $

Finalmente, la expresión intervalar para los cortes alfa de $ A^2 $ es:

$
A^2(\alpha) = \left[ \left( 2 - 2 \ln(1 - \alpha) \right)^2, \left( 2 + \ln(1 - \alpha) \right)^2 \right]
$

#### Conclusión

La expresión intervalar para los cortes alfa de $ A^2 $ está dada por los extremos izquierdo y derecho $ A_l^2(\alpha) $ y $ A_r^2(\alpha) $, como se muestra en la fórmula anterior.

### **Ejercicio 3.** Escribe una expresión para la función de membresía y los cortes alfa para un número difuso cuadrático $\text{qfn}(\bar x,\beta_l,\beta_r)$ en forma general. Utiliza este resultado para:

**1. Calcular los cortes alfa en forma intervalar del número $\text{qfn}(\bar x,\beta_l,\beta_r)$.**

````python
# Definimos las variables simbólicas
x, y, z, l, r = SR.var('x, y, z, l, r')  # indicamos cuáles son las variables

# Definimos la ecuación de la izquierda
ec_izquierda = y == 1 - (x - z)**2 / l**2  
ec_izquierda = solve(ec_izquierda, x)  # despejamos x

# Definimos la ecuación de la derecha
ec_derecha = y == 1 - (x - z)**2 / r**2  
ec_derecha = solve(ec_derecha, x)  # despejamos x
````

Tenemos así que el extremo izquierdo del corte alfa está dado por la siguiente ecuación:

ec_izquierda[0]:

$$
x = - l \cdot \frac{1}{-y + 1} + z

El corte alfa del lado derecho está dado por la siguiente ecuación:

ec_derecha[0]:

$$
x = r \cdot \frac{1}{-y + 1} + z
$$


En resumen tenemos que:

$$
\alpha_A = \left[ - \beta_l \cdot \frac{1}{-\alpha + 1} + \bar{x}, \, \beta_r \cdot \frac{1}{-\alpha + 1} + \bar{x} \right]
$$


**2. Calcular la función de membresía del número $\text{qfn}(\bar x,\beta_l,\beta_r)$.

$$
Q(x) = 
\begin{cases} 
    0, & \text{si } x \leq \bar{x} - \beta_l \\
    1 - \left(\frac{x - \bar{x}}{\beta_l}\right)^2, & \text{si } \bar{x} - \beta_l < x < \bar{x} \\
    1 - \left(\frac{x - \bar{x}}{\beta_r}\right)^2, & \text{si } \bar{x} \leq x < \bar{x} + \beta_r \\
    0, & \text{si } x \geq \bar{x} + \beta_r
\end{cases}
$$

**3. Presenta un dibujo de $\text{qfn}(1,-5,2)$.**

````python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la función difusa
x_barra = 1
beta_l = -5
beta_r = 2

# Definir la función qfn para los cortes alfa
def qfn(alpha, x_barra, beta_l, beta_r):
    # Cálculo de los extremos izquierdo y derecho del corte alfa
    izquierda = beta_l * np.sqrt(-alpha + 1) + x_barra
    derecha = beta_r * np.sqrt(-alpha + 1) + x_barra
    return izquierda, derecha

# Valores de alfa (entre 0 y 1 para la función difusa)
alpha_values = np.linspace(0, 1, 200)

# Inicializamos los arrays para los cortes alfa
izquierda_values = []
derecha_values = []

# Calcular los valores para cada alfa
for alpha in alpha_values:
    izquierda, derecha = qfn(alpha, x_barra, beta_l, beta_r)
    izquierda_values.append(izquierda)
    derecha_values.append(derecha)

# Graficar los cortes alfa
plt.fill_betweenx(alpha_values, izquierda_values, derecha_values, color="lightblue", label="qfn(1, -5, 2)")
plt.plot(izquierda_values, alpha_values, color="blue", label="Extremo Izquierdo")
plt.plot(derecha_values, alpha_values, color="green", label="Extremo Derecho")

# Configurar la gráfica
plt.title("Gráfico de $\\text{qfn}(1, -5, 2)$")
plt.xlabel("x")
plt.ylabel("Alfa")
plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()
````

Finalmente, $ u \cdot v $ se puede representar como un número difuso trapezoidal por el cuádruple $ (a, b, c, d) $.

La función de membresía de $ u \cdot v $ es igual a:

El intervalo del corte alfa de $ u \cdot v $ se define como:

$

a &= \min(2 \cdot -3, 2 \cdot 6, 8 \cdot -3, 8 \cdot 6) = \min(-6, 12, -24, 48) = -24 \\
